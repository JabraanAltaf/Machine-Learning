```python
# Importing pandas and numpy
import pandas as pd
import numpy as np
```

```python
# Reading the data set in using pandas
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/haberman/haberman.data',header=None)
```

```python
# Assigning values of features and responses to X and y
X,y = df.values[:,:3] ,df.values[:, 3]
print (X)
```

```python
# Printing out the shape of X and y
print (X.shape)
print (y.shape)
```

```python
# Importing metrics function
from sklearn import metrics
```

```python
# Splitting data into training set and test set using size of 0.4 - 40% for testing, 60% for training.
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=4)
# Initializing Logisitic Regression algorithm using default parameters
logreg = LogisticRegression()
logreg.fit(X_train, y_train) # Fiting model with training data set
```

```python
# Predicting on the test set using logistic regression
y_pred = logreg.predict(X_test)
#Comparing actual value response value(y_test) and predicted response value (y_pred)
print (metrics.accuracy_score(y_test, y_pred))
```

```python
# Making a confusion matrix using actual response and predicted response values
y = metrics.confusion_matrix(y_test, y_pred)

# Initliazing values from matrix to variables

#True Positive
TP = y[0,0] 
#False Negative 
FN = y[1,0]
# False Positive
FP = y[0,1]
#True Negative
TN = y[1,1]

#Printing out the Confusion Matrix
print (metrics.confusion_matrix(y_test, y_pred))
```

```python
# Working out the Accuracy Score
accuracy =  ((TN + TP)/(TN + FN + FP + TP))
# Working out the Error
error = 1 - accuracy
# Working out the True Positive Rate (Recall or Sensitivity)
true = TP/(TP + FN)
# Working out the False Positive Rate (Precision)
false = (TP/(FP + TP))
```

```python
#Printing out the scores from above
print ("Accuracy Score:", accuracy)
print ("Error:", error)
print ("True Positive:", true)
print("False Positive:", false)
```
